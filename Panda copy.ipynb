# Filtering and ordering panda dataframes

1. Initialise and reading file path

```python
import pandas as pd
```

```python
df = pd.read_csv(r"/Users/mattmeredith/Downloads/world-data-2023.csv")
```

2. Filter by single condition or specific values

```python
df[df['Fertility Rate'] >= 5]
```

```python
specificCountries = ['Australia', 'India']
```

```python
df[df['Country'].isin(specificCountries)]
```

```python
df[df['Country'].str.contains('Island')]
```

3. Setting index of a dataframe

```python
df2 = df.set_index('Abbreviation')
```

4. Filtering dataframe with .filter() and like

```python
df2.filter(like = 'X', axis = 0)
```

5. Locating specific data with loc[] and specific index position with iloc[]

```python
df2.loc['MX']
```

```python
df.iloc[6]
```

6. Order by .sort_values()

```python
df[df['Fertility Rate'] > 5].sort_values(by='Birth Rate', ascending=False)
```
